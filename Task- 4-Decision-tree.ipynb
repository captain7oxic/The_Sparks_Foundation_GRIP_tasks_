# Exploring Decision Trees

### Importing Necessary Libraries

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

### Loading Iris Data through scikit 

```python
data=load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
```

```python
df.head(10)
```

### splitting the data into training and test sets.

```python
X_train, X_test, Y_train, Y_test = train_test_split(df[data.feature_names], df['target'], random_state=0)
```

### Defing the model

```python
clf = DecisionTreeClassifier(max_depth = 5,random_state = 0)
```

#### Fitting the Training Data to the Model

```python
clf.fit(X_train,Y_train)
```

```python
from sklearn.tree import plot_tree as pt
```

```python
feature_names_list=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
species_list=['setosa', 'versicolor', 'virginica']
```

```python
pt(clf);
```

## Displaying the plot

```python
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=300)
pt(clf,feature_names = feature_names_list, 
               class_names=species_list,
               filled = True)
```

### Predicting the Output

```python
test=[[5.0,3.5,4.2,0.5]]
```

```python
clf.predict(test)
```
